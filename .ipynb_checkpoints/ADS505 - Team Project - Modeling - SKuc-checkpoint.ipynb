```python
# importing necessary libraries

import numpy as np
import pandas as pd

from sklearn import preprocessing
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV, LinearRegression
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
from sklearn.metrics import  classification_report, confusion_matrix, roc_curve
from sklearn.neighbors import NearestNeighbors, KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

from imblearn.over_sampling import SMOTE

import matplotlib.pylab as plt
import seaborn as sns

import dmba
from dmba import classificationSummary, gainsChart, liftChart, regressionSummary, stepwise_selection, plotDecisionTree
from dmba.metric import AIC_score, adjusted_r2_score, BIC_score
```

```python
# Filtering warnings
import warnings
warnings.filterwarnings("ignore")
```

```python
# re-loading in churn_df after pre-processing, scaling, etc. from EDA and Pre-processing code file

churn_df = pd.read_csv("C:/Users/steph/OneDrive/Documents/USD/ADS505/Data/churn_final.csv", index_col = 0)
```

```python
churn_df.shape
```

```python
# splitting into X and Y

X = churn_df.drop('churn', axis = 1)
y = churn_df['churn']


# Balancing data


smote = SMOTE(sampling_strategy='minority',random_state=0)
X_smt, y_smt = smote.fit_resample(X,y)

y_smt.value_counts()

```

```python
# partitioning data into 90% train and 10% test, and then will further split for validation

trainX, testX, trainy, testy = train_test_split(X_smt, y_smt, test_size = .1, random_state = 12345)

# using 80 / 20 split for train and validation set

trainX, validX, trainy, validy = train_test_split(trainX, trainy, test_size = .2, random_state = 12345)
```

```python
print("Test set dimensions: ", testX.shape, 
      "\n Train set dimensions: ", trainX.shape,
      "\n Validation set dimension: ", validX.shape)
```

# Modeling

### K-Nearest Neighbors

```python
# first, k-nn

# Train a classifier for different values of k
knn_results = []
for k in range(1, 25, 2):
    knn = KNeighborsClassifier(n_neighbors=k).fit(trainX, trainy)
    valid_preds = knn.predict(validX)
    knn_results.append({
        'k': k,
        'accuracy': accuracy_score(validy, valid_preds),
        'precision': precision_score(validy, valid_preds ),
        'recall': recall_score(validy, valid_preds),
        'f1_score': f1_score(validy, valid_preds)
    })

# Convert results to a pandas data frame
knn_results = pd.DataFrame(knn_results)
knn_results
```

```python
# Plotting K-nn metrics from validation tests

# show accuracy by k
plt.plot(knn_results.k, knn_results.accuracy, linestyle='-', c='green', label = "accuracy")
plt.plot(knn_results.k, knn_results.recall, linestyle=':', c='red', label = "recall")
plt.plot(knn_results.k, knn_results['f1_score'], linestyle='--', c='blue', label = "f1 score")
plt.legend()
plt.title('K vs. Metrics in validation set')
plt.xlabel('K')
plt.ylabel('Metrics score')
```

```python
# fitting using optimal parameters of k = 3

knn = KNeighborsClassifier(n_neighbors=3).fit(trainX, trainy)

# Prediction using Test dataset
predictions_knn = knn.predict(testX)
probabilities_knn = knn.predict_proba(testX)[:,1]

knn_result = pd.DataFrame({'actual': testy, 
                             'probability': probabilities_knn,
                             'predicted': predictions_knn})

knn_result
```

```python
print ("\n Classification report for K-NN : \n",classification_report(testy,predictions_knn))
print ("Accuracy Score : ",accuracy_score(testy,predictions_knn))


#confusion matrix
conf_matrix = confusion_matrix(testy,predictions_knn)
plt.figure(figsize=(12,12))
plt.subplot(221)
sns.heatmap(conf_matrix, fmt = "d",annot=True, cmap='Blues')
plt.title('Confuion Matrix for KNN')
plt.ylabel('True Values')
plt.xlabel('Predicted Values')


#roc_auc_score
model_roc_auc = roc_auc_score(testy,probabilities_knn) 
print ("Area under curve : ",model_roc_auc,"\n")
fpr,tpr,thresholds = roc_curve(testy,probabilities_knn)

plt.subplot(222)
plt.plot(fpr, tpr, color='darkorange', lw=1, label = "Auc : %.3f" %model_roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC')
plt.legend(loc="lower right")

    
plt.show()
```

### Logistic Regression

```python
grid_parameters = { 
    'solver': ['newton-cg', 'lbfgs', 'liblinear'],
    'penalty': ['l1', 'l2'],
    'C' : np.logspace(-4, 4, 20)
}
logit = GridSearchCV(LogisticRegression(random_state=42), grid_parameters, cv= 5, scoring = "recall")
logit = logit.fit(trainX, trainy)
logit_final = logit.best_estimator_
```

```python
# seeing best parameters of final model
logit.best_params_
```

```python
# seeing coefficients for final logistc regression model

print(pd.DataFrame({'coeff': logit_final.coef_[0]}, index=X.columns).transpose())
```

It appears that this is only using age as a coefficient

```python
# validation set 

#Confusion matrix
cm_valid_lr =confusion_matrix(validy,logit_final.predict(validX))
print(cm_valid_lr)
sns.heatmap(cm_valid_lr,annot=True)

print(classification_report(validy,logit_final.predict(validX)))

```

#### Logistic Regression - Test Results

```python
predictions_lr = logit_final.predict(testX)
probabilities_lr = logit_final.predict_proba(testX)[:,1]

logit_result = pd.DataFrame({'actual': testy, 
                             'probability': probabilities_lr,
                             'predicted': predictions_lr})

logit_result
```

```python
# Logistic regression Test results

print ("\n Classification report for Logistic Regression \: \n",classification_report(testy,predictions_lr))
print ("Accuracy Score : ",accuracy_score(testy,predictions_lr))
print("Recall: ", recall_score(testy, predictions_lr))


#confusion matrix
conf_matrix = confusion_matrix(testy,predictions_lr)
plt.figure(figsize=(12,12))
plt.subplot(221)
sns.heatmap(conf_matrix, fmt = "d",annot=True, cmap='Blues')
plt.title('Confuion Matrix for Logistic Regression')
plt.ylabel('True Values')
plt.xlabel('Predicted Values')


#roc_auc_score
model_roc_auc = roc_auc_score(testy,probabilities_lr) 
print ("Area under curve : ",model_roc_auc,"\n")
fpr,tpr,thresholds = roc_curve(testy,probabilities_lr)

plt.subplot(222)
plt.plot(fpr, tpr, color='darkorange', lw=1, label = "Auc : %.3f" %model_roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC')
plt.legend(loc="lower right")

    
plt.show()
```

### Decision Tree

```python
# decision tree

dt_results = []
criterion = ["gini", "entropy", "log_loss"]


for x in criterion:
    for y in range(1,15,1):
        dt = DecisionTreeClassifier(criterion = x, max_depth = y, random_state = 1).fit(trainX, trainy)
        valid_preds = dt.predict(validX)
        dt_results.append({'criterion': x, 'max_depth': y,
                           'accuracy': accuracy_score(validy, valid_preds), 
                      'precision': precision_score(validy, valid_preds),
                           'recall': recall_score(validy, valid_preds),
                      'f1_score': f1_score(validy, valid_preds)})
    

# Convert results to a pandas data frame
dt_results = pd.DataFrame(dt_results)
```

```python
dt_results
```

```python
# show accuracy by depth
plt.plot(dt_results['max_depth'], dt_results.accuracy, linestyle='-', label = "accuracy")
plt.plot(dt_results['max_depth'], dt_results.recall, linestyle=':', c='red', label = "recall")
plt.plot(dt_results['max_depth'], dt_results['f1_score'], linestyle='--', c='blue', label = "f1 score")
plt.title('Number of estimators vs. metrics \n In validation set for decision tree \n Using Entropy Loss')
plt.legend()
plt.xlabel('Max depth')
plt.ylabel('Metrics score')
```

Looks like with max depth around 12, and gini index (recall is slightly lower but the rest of the metrics are higher), those are the optimal tuning parameters. 

```python
# Prediction using Test dataset with Decision tree
dt_final = DecisionTreeClassifier(criterion = 'gini', max_depth = 12, random_state = 1).fit(trainX, trainy)
predictions_dt = dt_final.predict(testX)
probabilities_dt = dt_final.predict_proba(testX)[:,1]

dt_result = pd.DataFrame({'actual': testy, 
                             'probability': probabilities_dt,
                             'predicted': predictions_dt})

dt_result
```

#### Decision Tree - Test Results

```python
# Decision Tree Test results

print ("\n Classification report for Decision Tree \: \n",classification_report(testy,predictions_dt))
print ("Accuracy Score : ",accuracy_score(testy,predictions_dt))
print("Recall: ", recall_score(testy, predictions_dt))


#confusion matrix
conf_matrix = confusion_matrix(testy,predictions_dt)
plt.figure(figsize=(12,12))
plt.subplot(221)
sns.heatmap(conf_matrix, fmt = "d",annot=True, cmap='Blues')
plt.title('Confuion Matrix for Decision Tree')
plt.ylabel('True Values')
plt.xlabel('Predicted Values')


#roc_auc_score
model_roc_auc = roc_auc_score(testy,probabilities_dt) 
print ("Area under curve : ",model_roc_auc,"\n")
fpr,tpr,thresholds = roc_curve(testy,probabilities_dt)

plt.subplot(222)
plt.plot(fpr, tpr, color='darkorange', lw=1, label = "Auc : %.3f" %model_roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC')
plt.legend(loc="lower right")

    
plt.show()
```

### Random Forest

```python
# random forest tuning hyperparameters with Grid Search
# Took very long to tune, longer than the below further tuning

grid_parameters = { 
    'n_estimators': range(100,1001,100),
    'max_depth' : range(4,12,1),
    'criterion' :['gini', 'entropy']
}


rf= GridSearchCV(RandomForestClassifier(random_state=42, max_features = 'sqrt'), grid_parameters, cv= 5)
rf=rf.fit(trainX, trainy)
rf_clf=rf.best_estimator_

```

```python
#best parameters for random forest classifier
rf.best_params_
```

```python
#Confusion matrix on validation set
valid_cm_rf =confusion_matrix(validy,rf_clf.predict(validX))
print(valid_cm_rf)
sns.heatmap(valid_cm_rf,annot=True)

print(classification_report(validy,rf_clf.predict(validX)))

```

```python
# random forest

# Train a classifier for different values of n_estimators and max_depth
rf_results = []
for x in range(100, 1001, 100):
    for y in range(1,12,1):
        rf = RandomForestClassifier(n_estimators = x, max_depth = y, random_state = 1).fit(trainX, trainy)
        valid_preds = rf.predict(validX)
        rf_results.append({'n_estimators': x, 'max_depth': y, 'accuracy': accuracy_score(validy, valid_preds), 
                      'precision': precision_score(validy, valid_preds), 
                           'recall': recall_score(validy, valid_preds),
                      'f1_score': f1_score(validy, valid_preds)})

    
# Convert results to a pandas data frame
rf_results = pd.DataFrame(rf_results)
rf_results
```

```python
rf_results['recall'].max()
```

```python
# show accuracy by estimators
plt.plot(rf_results['n_estimators'], rf_results.accuracy, linestyle='-', label = "accuracy")
plt.plot(rf_results['n_estimators'], rf_results.recall, linestyle=':', c='red', label = "Recall")
plt.plot(rf_results['n_estimators'], rf_results['f1_score'], linestyle='--', c='blue', label = "f1 score")
plt.title('Number of estimators vs. metrics \n In validation set for Random Forest')
plt.legend()
plt.xlabel('Estimators')
plt.ylabel('Metrics score')
```

```python
# show accuracy by max_depth
plt.plot(rf_results['max_depth'], rf_results.accuracy, linestyle='-', label = "accuracy")
plt.plot(rf_results['max_depth'], rf_results.recall, linestyle=':', c='red', label = "recall")
plt.plot(rf_results['max_depth'], rf_results['f1_score'], linestyle='--', c='blue', label = "f1 score")
plt.title('Number of estimators vs. metrics \n In validation set for Random Forest')
plt.legend()
plt.xlabel('Estimators')
plt.ylabel('Metrics score')
```
